```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
df = pd.read_csv('Wojood/Wojood1_1_flat/train.csv')
df.head()
```

```python
df = df[['global_sentence_id', 'token', 'tags']]
df.head()
```

```python
words = list(set(df["token"].values))

tags = list(set(df["tags"].values))
num_tags = len(tags)
```

```python
def sentence_integrate(data):
  agg_func = lambda s: [(w, t) for w, t in zip(s["token"].values.tolist(),
                                                           s["tags"].values.tolist())]
  return data.groupby('global_sentence_id').apply(agg_func).tolist()
  
sentences=sentence_integrate(df)
```

```python
sentences[1]
```

```python
word2idx = {w: i + 2 for i, w in enumerate(words)}
tag2idx = {t: i for i, t in enumerate(tags)}
```

```python
word2idx["PAD"] = 0
word2idx["UNK"] = 1
```

```python
word2idx
```

```python
idx2tag = {i: t for t, i in tag2idx.items()}
idx2tag
```

```python
import pickle

with open("word2idx.pkl", "wb") as f:
    pickle.dump(word2idx, f)

with open("tag2idx.pkl", "wb") as f:
    pickle.dump(tag2idx, f)

with open("idx2tag.pkl", "wb") as f:
    pickle.dump(idx2tag, f)
```

```python
num_words = len(word2idx)
num_words
```

```python
max(len(s) for s in sentences)
```

```python
min(len(s) for s in sentences)
```

```python
max_len = 135
X = [[word2idx[w[0]] for w in s] for s in sentences]
X = pad_sequences(maxlen=max_len, sequences=X, padding="post", value=num_words-1)

y = [[tag2idx[w[1]] for w in s] for s in sentences]
y = pad_sequences(maxlen=max_len, sequences=y, padding="post", value=tag2idx["O"])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

#build model
from tensorflow.keras.layers import LSTM, Embedding, InputLayer, SpatialDropout1D, Bidirectional
from tensorflow import keras

model = keras.Sequential()
model.add(InputLayer(input_shape=(max_len,)))
model.add(Embedding(input_dim=num_words, output_dim=max_len, input_length=max_len))
model.add(SpatialDropout1D(0.1))
model.add(Bidirectional(LSTM(units=100, return_sequences=True, recurrent_dropout=0.1)))


model.summary()


model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
history = model.fit(
    x_train, 
    np.array(y_train), 
    batch_size=32, 
    epochs=5, 
    validation_split=0.1, 
    verbose=1
)
```

```python
model.save('model.keras')
```
